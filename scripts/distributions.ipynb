```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/test.csv', encoding='unicode_escape', sep=';', header=None)
data.columns = ['Type', 'Month', 'Total', 'National', 'International']
```

```python
data = data[data['Type'] == 'Flyvninger i alt']
```

```python
national_counts, national_bins = np.histogram(data['National'].dropna(), bins=10)
international_counts, international_bins = np.histogram(data['International'].dropna(), bins=10)

plt.hist(national_bins[:-1], national_bins, weights=national_counts, edgecolor='black', label='National', alpha=0.5)
plt.hist(international_bins[:-1], international_bins, weights=international_counts, edgecolor='black', label='International', alpha=0.5)
plt.xlabel('Passengers per month (thousands)')
plt.ylabel('Count (months)')
plt.title(f'Distribution of passengers per month')
plt.grid()
plt.legend()
plt.show()
```
